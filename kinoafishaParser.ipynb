```python
import requests
from bs4 import BeautifulSoup
```

```python
letters = [chr(i) for i in range(ord('А'), ord('Я')+1)]
letters.append('A-Z')
letters.append('0-9%23')
urls = set()
for i in letters:
    url = f'https://www.kinoafisha.info/movies/archive/letter/{i}/'
    response = requests.get(url)
    if response.status_code != 200:
        print('Well')
    soup = BeautifulSoup(response.text, features='html.parser')
    des = soup.find_all("a", class_="archiveList_movie")
    for t in des:
        urls.add(t.get('href')+'\n')
print(urls)
```
